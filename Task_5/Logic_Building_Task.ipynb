```python
#1 – Social Media Post Engagement Analyzer

def analyze_engagement(likes_list):
    """Calculates total likes using a loop and determines if the post is viral."""
    total_likes = 0
    for like in likes_list:
        total_likes += like
        
    status = "Viral Post" if total_likes >= 1000 else "Normal Engagement"
    
    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {status}")

# Test Case
analyze_engagement([200, 300, 550])
```

```python
#2 - Healthcare Medicine Stock Alert System

def monitor_stock(stock_level):
    """Checks if medicine stock is sufficient or needs an alert."""
    if stock_level < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"
        
    print(f"Medicine Stock: {stock_level}")
    print(f"Status: {status}")

# Test Case
monitor_stock(6)
monitor_stock(15)
```

```python
#3 - Agriculture Rainfall Adequacy Checker

def check_rainfall(rainfall_data):
    """Calculates average rainfall and determines adequacy."""
    required_level = 70  # Example threshold based on expected output
    total_rainfall = 0
    
    for rain in rainfall_data:
        total_rainfall += rain
    
    avg_rainfall = total_rainfall / len(rainfall_data)
    status = "Adequate Rainfall" if avg_rainfall >= required_level else "Inadequate Rainfall"
    
    print(f"Average Rainfall: {avg_rainfall:.0f}")
    print(f"Rainfall Status: {status}")

# Test Case
check_rainfall([80, 65, 71, 72])
check_rainfall([50, 60, 65, 45])
```

```python
#4 – Social Media Duplicate Account Detection

def detect_duplicates(usernames):
    """Uses a set to identify if duplicate usernames exist in the list."""
    # A set only contains unique items; if length changes, duplicates existed
    if len(usernames) != len(set(usernames)):
        result = "Yes"
    else:
        result = "No"
        
    print(f"Duplicate Accounts Found: {result}")

# Test Case
detect_duplicates(["user1", "user2", "user1", "user3"])
detect_duplicates(["user1", "user2", "user4", "user3"])
```

```python
#5 – Healthcare Appointment Eligibility Checker

def check_eligibility(age):
    """Checks if a patient meets the age requirement for an appointment."""
    status = "Eligible" if age >= 18 else "Not Eligible"
    
    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

# Test Case
check_eligibility(21)
check_eligibility(12)
```

```python
#6 – Agriculture Premium Crop Price Filter

def filter_premium_crops(price_list):
    # 1. Create an empty list to store premium prices
    premium_crops = []
    
    # 2. Use a loop to check each price
    for price in price_list:
        # 3. Apply the rule: Price must be greater than 2000
        if price > 2000:
            premium_crops.append(price)
            
    # 4. Print the result in the expected format
    print(f"Premium Crops: {premium_crops}")

# --- Test the function ---
prices = [1500, 2500, 1800, 3200]
filter_premium_crops(prices)
```

```python
#7 – System Monitoring Application Health Checker

def check_app_health(error_count):
    """Evaluates system health based on the number of recorded errors."""
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"
        
    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Test Case
check_app_health(7)
```

```python
#8 – Banking Daily Transaction Limit Checker

def validate_transaction(amount):
    """Validates if a transaction amount stays within the daily limit of 50,000."""
    limit = 50000
    status = "Approved" if amount <= limit else "Rejected"
    
    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# Test Case
validate_transaction(60000)
```

```python
#9 – E-Learning Student Attendance Eligibility System

def calculate_attendance(attendance_list):
    """Calculates attendance percentage and determines exam eligibility."""
    # Assuming list contains 1 for present and 0 for absent
    total_days = len(attendance_list)
    days_present = sum(attendance_list)
    
    percentage = (days_present / total_days) * 100
    status = "Eligible" if percentage >= 75 else "Not Eligible"
    
    print(f"Attendance Percentage: {percentage:.1f}")
    print(f"Exam Eligibility: {status}")

# Test Case
calculate_attendance([1, 1, 1, 0, 1]) # 4 out of 5 = 80%
```

```python
#10 – Smart Electricity Bill Analyzer

def analyze_electricity_bill(units):
    """Calculates bill based on tiered slabs and classifies usage intensity."""
    bill = 0
    
    # Calculate Bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7
        
    # Classify Usage
    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"
        
    print(f"Units Consumed: {units}")
    print(f"Total Bill: ₹{bill}")
    print(f"Usage Classification: {usage_status}")

# Test Case
analyze_electricity_bill(250)
```
