```python
#1.Unique Words in a Sentence

# Input sentence
sentence = "python is easy and python is powerful"

# Step 1: Split sentence into a list of words
words = sentence.split()

# Step 2: Convert list to a set to filter duplicates
unique_words = set(words)

# Output
print(f"Unique words count: {len(unique_words)}")
print(f"Unique words: {unique_words}")
```

```python
#2.Highest Salary from Employee Data

employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Identify the key (name) with the maximum value (salary)
highest_paid = max(employees, key=employees.get)

print(f"Highest Salary: {highest_paid} - {employees[highest_paid]}")
```

```python
#3.Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]

# Initialize with the first element
max_val = numbers[0]
min_val = numbers[0]

# Iterate and compare
for num in numbers:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print(f"List: {numbers}")
print(f"Max: {max_val}\nMin: {min_val}")
```

```python
#4.Products Above Price Threshold

prices = [450, 1200, 899, 1500, 300]
count = 0

for p in prices:
    if p > 1000:
        count += 1

print(f"Products above 1000: {count}")
```

```python
#5.Attendance Percentage

attendance = ["P", "P", "A", "P", "P"]

# Count 'P' for Present
present_count = attendance.count("P")
total_days = len(attendance)

# Formula: (Part / Total) * 100
percentage = (present_count / total_days) * 100

print(f"Attendance Percentage: {percentage}")
```

```python
#6.Remove Duplicate Phone Numbers

phone_numbers = [9876543210, 9123456789, 9876543210]

# Sets are ideal for removing duplicates in CRM or marketing data
unique_phones = set(phone_numbers)

print(f"Unique phone numbers: {unique_phones}")
```

```python
#7.Character Frequency Count

text = "pythonp"
freq = {}

for char in text:
    # If char exists, increment; else start at 0 and add 1
    freq[char] = freq.get(char, 0) + 1

print(freq)
```

```python
#8.Convert List to Tuple

my_list = [10, 20, 30]
my_tuple = tuple(my_list)

print(f"Tuple: {my_tuple}")
```

```python
#10.Calculate Average Marks

marks = [70, 80, 65, 75]

# sum() / len() is the most efficient way to find the mean
avg = sum(marks) / len(marks)

print(f"Average Marks: {avg}")
```
