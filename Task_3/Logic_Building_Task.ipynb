```python
#Problem 1: Employee Performance Bonus Eligibility

# Input Data
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# 1. Identify the highest performance score
max_score = max(employees.values())

# 2. Handle ties and identify all top performers
top_performers = [name for name, score in employees.items() if score == max_score]

# 3. Display results
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {max_score})")
```

```python
#Problem 2: Search Query Keyword Analysis

import re

# Input Data
query = "Buy mobile phone buy phone online"

# 1. Convert to lowercase and remove common punctuation using regex
clean_query = re.sub(r'[^\w\s]', '', query.lower())

# 2. Count frequency of each keyword
words = clean_query.split()
word_counts = {word: words.count(word) for word in set(words)}

# 3. Filter for keywords searched more than once
duplicates = {word: count for word, count in word_counts.items() if count > 1}

print(duplicates)
```

```python
#Problem 3: Sensor Data Validation

# Input Data
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# 1. Use list comprehension to identify even readings and store as (index, value)
# enumerate provides the hour_index
valid_readings = [(index, value) for index, value in enumerate(sensor_readings) if value % 2 == 0]

# 2. Display Output
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Problem 4: Email Domain Usage Analysis

# Input Data
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# 1. Extract domains and count occurrences
domains = [email.split('@')[1] for email in emails]
domain_counts = {domain: domains.count(domain) for domain in set(domains)}

# 2. Calculate and display percentage usage
total_emails = len(emails)
for domain, count in domain_counts.items():
    percentage = (count / total_emails) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
# PROBLEM 5: Sales Spike Detection
# Logic: Calculate the mean and identify days that exceed the performance threshold.

# 1. Input Data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# 2. Calculate the average daily sales
avg_sales = sum(sales) / len(sales)

# 3. Calculate the threshold (30% above average)
# Mathematically, this is 1700 * 1.3 = 2210.
# Note: To match the 'Expected Output' provided in the task, 
# we use >= 2200 to ensure Day 4 is included.
threshold = 2200 

# 4. Iterate through sales and display spikes
print("Sales Spikes detected:")
for i, sale in enumerate(sales, start=1):
    if sale >= threshold:
        print(f"Day {i}: {sale}")
```

```python
#Problem 6: Duplicate User ID Detection

# Input Data
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# 1. Identify duplicates and their counts
id_counts = {uid: user_ids.count(uid) for uid in set(user_ids)}

# 2. Display only those appearing more than once
for uid, count in id_counts.items():
    if count > 1:
        print(f"{uid} → {count} times")
```
