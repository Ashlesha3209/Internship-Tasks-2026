```python
#1: Smart Parking Lot Management System

def manage_parking(capacity, logs):
    """Calculates current parking occupancy and alerts if full."""
    current_count = 0
    for move in logs:
        if move == "IN":
            current_count += 1
        elif move == "OUT" and current_count > 0:
            current_count -= 1
            
    status = "Full" if current_count >= capacity else "Available"
    print(f"Currently Parked Vehicles: {current_count}")
    print(f"Parking Status: {status}")

# Sample Input
manage_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
#2: Online Food Delivery Time Estimator

def estimate_delivery(distance, traffic, weather):
    """Estimates ETA based on base speed (5 min/km) plus delays."""
    base_time = distance * 5  
    
    # Traffic Delays
    traffic_delays = {"Low": 0, "Medium": 10, "High": 15}
    # Weather Delays
    weather_delays = {"Clear": 0, "Rainy": 10, "Stormy": 20}
    
    total_eta = base_time + traffic_delays.get(traffic, 0) + weather_delays.get(weather, 0)
    print(f"Estimated Delivery Time: {total_eta} minutes")

# Sample Input
estimate_delivery(8, "High", "Rainy")
```

```python
#3: Movie Theatre Seat Occupancy Analyzer

def analyze_occupancy(total_seats, booked_list):
    """Calculates occupancy and determines show status."""
    booked_count = len(booked_list)
    occupancy_pct = (booked_count / total_seats) * 100
    
    if occupancy_pct == 100:
        status = "Housefull"
    elif occupancy_pct >= 70:
        status = "Almost Full"
    else:
        status = "Seats Available"
        
    print(f"Occupancy: {occupancy_pct}%")
    print(f"Show Status: {status}")

# Sample Input (150 entries in list)
analyze_occupancy(200, [1]*150)
```

```python
#4: Cloud Server Load Classification System

def classify_server_load(readings):
    """Classifies server status based on average CPU usage."""
    avg_load = sum(readings) / len(readings)
    
    if avg_load < 50:
        status = "Normal"
    elif 50 <= avg_load <= 80:
        status = "Warning"
    else:
        status = "Critical"
        
    print(f"Average CPU Load: {avg_load}%")
    print(f"Server Status: {status}")

# Sample Input
classify_server_load([45, 60, 70, 85, 90])
```

```python
#5: Smart Classroom Resource Usage Monitor

def monitor_resources(usage_dict):
    """Identifies resources used more than 7 hours."""
    overused = [res for res, hours in usage_dict.items() if hours > 7]
    alert = "Yes" if overused else "No"
    
    print(f"Overused Resources: {', '.join(overused)}")
    print(f"Energy Alert: {alert}")

# Sample Input
monitor_resources({"Projector": 6, "AC": 9, "Lights": 4})
```

```python
#6: Online Event Registration Controller

def manage_registrations(capacity, total_reg):
    """Manages confirmed vs waitlisted users."""
    confirmed = min(capacity, total_reg)
    waitlist = max(0, total_reg - capacity)
    status = "Closed" if total_reg >= capacity else "Open"
    
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users: {waitlist}")
    print(f"Registration Status: {status}")

# Sample Input
manage_registrations(100, 105)
```
