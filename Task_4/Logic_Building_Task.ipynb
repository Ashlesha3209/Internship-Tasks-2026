```python
#1. Mobile Recharge Validation System

def validate_recharge():
    valid_plans = [199, 299, 399, 599]
    
    while True:
        try:
            amount = int(input("Enter recharge amount: "))
            
            # Rule 1: Minimum amount check
            if amount < 50:
                print(f"Error: Minimum recharge amount is ₹50.")
            # Rule 2: Plan validity check
            elif amount not in valid_plans:
                print(f"Error: {amount} is not a valid plan. Choose from {valid_plans}.")
            else:
                print(f"Success! Recharge of ₹{amount} is successful.")
                break # Exit loop on success
                
        except ValueError:
            print("Invalid input. Please enter a numeric value.")
        
        # Rule 3: Retry logic
        retry = input("Try again? (yes/no): ").lower()
        if retry != 'yes':
            break

validate_recharge()
```

```python
#2. Inventory Reorder Alert System

def check_inventory(stock_dict):
    # Iterate through dictionary using items()
    for product, quantity in stock_dict.items():
        if quantity < 15:
            print(f"Reorder Alert: {product} (Stock: {quantity})")
        else:
            print(f"Stock OK: {product} (Stock: {quantity})")

# Sample Data
inventory = {"Laptop": 20, "Mouse": 10, "Keyboard": 25, "Monitor": 8}
check_inventory(inventory)
```

```python
#3. Student Result Processing System

def process_result(marks_list):
    if not marks_list: return "No data"
    
    total = 0
    # Loop to calculate total
    for mark in marks_list:
        total += mark
        
    average = total / len(marks_list)
    
    # Logic to determine pass/fail
    if average >= 50:
        return f"Pass (Average: {average:.2f})"
    else:
        return f"Fail (Average: {average:.2f})"

# Sample Execution
marks = [65, 40, 78, 32, 55]
print(process_result(marks))
```

```python
#4. Cab Fare Estimator with Retry Option

def calculate_fare(km, peak_hour):
    base_fare = 50
    rate_per_km = 12
    total = base_fare + (km * rate_per_km)
    
    if peak_hour:
        total *= 1.25 # Adding 25% peak hour charge
    return total

while True:
    try:
        dist = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ").lower() == 'yes'
        
        fare = calculate_fare(dist, peak)
        print(f"Estimated Fare: ₹{fare:.2f}")
    except ValueError:
        print("Please enter a valid number for distance.")
        
    if input("New calculation? (yes/no): ").lower() != 'yes':
        break
```

```python
#5. Employee Attendance Eligibility Checker

def check_eligibility(attendance):
    present_days = 0
    # Count present days using a loop
    for status in attendance:
        if status == "P":
            present_days += 1
            
    percentage = (present_days / len(attendance)) * 100
    
    # Return eligibility status
    if percentage >= 75:
        return f"Eligible ({percentage:.1f}%)"
    else:
        return f"Not Eligible ({percentage:.1f}%)"

# Sample Data
attendance_log = ["P", "P", "A", "P", "P", "P", "A", "P"]
print(check_eligibility(attendance_log))
```

```python
#6. Password Strength Checker

def check_password_strength(password):
    # Rule 1: Length Check
    if len(password) < 8:
        return "Weak: Password must be at least 8 characters long."
    
    has_digit = False
    has_special = False
    special_chars = "@#$"
    
    # Loop over string to check each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in special_chars:
            has_special = True
            
    # Final Validation
    if has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak: Must contain a digit and a special character (@, #, or $)."

# Test Case
print(check_password_strength("User@123"))
```
