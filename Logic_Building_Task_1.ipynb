```python
#1.User Login Check

# Pre-defined credentials
username_stored = "admin"
password_stored = "1234"

# Simulating the login inputs
input_username = "admin"
input_password = "1234"

# Logic: Both conditions must be True for a successful login
if input_username == username_stored and input_password == password_stored:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
#2.Pass / Fail Analyzer

# List of student marks to analyze
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Iterating through the list to check each student's score
for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Output the final counts
print(f"Total pass students: {pass_count}")
print(f"Total fail students: {fail_count}")
```

```python
#3.Simple Data Cleaner

# Raw list with inconsistent formatting
names = [" Alice ", "bob", " CHARLIE "]

# Using list comprehension to strip spaces and convert to lowercase
# .strip() removes leading/trailing spaces; .lower() standardizes text
cleaned_names = [name.strip().lower() for name in names]

print("Cleaned List:", cleaned_names)
```

```python
#4.Message Length Analyzer

# List of messages for size analysis
messages = ["Hi", "Welcome to the platform", "OK"]

for msg in messages:
    msg_length = len(msg)
    
    # Check if the message exceeds the 10-character limit
    if msg_length > 10:
        print(f"Message: '{msg}' | Length: {msg_length} | Status: FLAG (Long Message)")
    else:
        print(f"Message: '{msg}' | Length: {msg_length} | Status: OK")
        
```

```python
#5.Error Message Detector

# System logs to be monitored
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Using the count method to find specific occurrences of "ERROR"
error_count = logs.count("ERROR")

print(f"Total error count: {error_count}")
```
