```python
#1: Smart Light Controller

class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"

    def set_status(self, action):
        self.status = action.upper()
        print(f"{self.name} is {self.status}")

# Usage
light = SmartLight("Bedroom Light")
light.set_status("ON")
```

```python
#2: Employee ID Card System

class Employee:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")

# Usage
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

```python
#3: Mobile Contact Record

class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")

# Usage
contact = Contact("Anita", "9876543210")
contact.display_contact()
```

```python
#4: Product Price Tag Generator

class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def print_price_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")

# Usage
prod = Product("Headphones", 2499)
prod.print_price_tag()
```

```python
#5: Movie Rating Display System

class Movie:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_details(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating} / 5")

# Usage
movie = Movie("Inception", 4.8)
movie.display_details()
```

```python
#6: Delivery Address Manager

class Delivery:
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address

    def display_delivery_details(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")

# Usage
order = Delivery("Suman", "Hyderabad")
order.display_delivery_details()
```
